# Project Overview
This project will use data from the World Health Organization and the World Bank to study possible correlations between Gross Domestic Product (GDP) and life expectancy in six countries.

**Questions To Be Answered**
- How has life expentancy changed over time?
- How has GDP changed over time?
- Is there a correlation between GDP and life expectancy?

**Countries Being Studied**
Chile - China - Germany - Mexico - USA - Zimbabwe

**Data Overview**
For each country, life expectancy at birth and GDP is given for each year from 2000 to 2015.

**Data Sources**
- GDP Source: [World Bank](https://data.worldbank.org/indicator/NY.GDP.MKTP.CD) national accounts data, and OECD National Accounts data files.

- Life expectancy Data Source: [World Health Organization](http://apps.who.int/gho/data/node.main.688)

**Create function to plot data with best fit line.**

```python
def plotBestFit(df,colX,colY,color='k',title='Title',xlab='x',ylab='y'):
    x = df[colX]
    y = df[colY]
    slope, intercept = np.polyfit(x,y,1)
    y_fit = slope * x + intercept

    #plt.figure()
    plt.scatter(x,y,color=color)
    plt.plot(x,y_fit,color='k',label='Slope: '+str(slope))
    plt.legend()
    plt.title(title)
    plt.ylabel(ylab)
    plt.xlabel(xlab)
    #plt.show()
```

Subplot of all countries with best fit line.

```python
plt.figure(figsize=(16,16))
plt.subplot(3,2,1)
plotBestFit(CL,'GDP','LE','b','GDP vs LE for Chile','GDP in Trillions of U.S. Dollars','Life Expectancy At Birth (year)')

plt.subplot(3,2,2)
plotBestFit(CN,'GDP','LE','r','GDP vs LE for China','GDP in Trillions of U.S. Dollars','Life Expectancy At Birth (year)')

plt.subplot(3,2,3)
plotBestFit(DE,'GDP','LE','g','GDP vs LE for Germany','GDP in Trillions of U.S. Dollars','Life Expectancy At Birth (year)')

plt.subplot(3,2,4)
plotBestFit(MX,'GDP','LE','c','GDP vs LE for Mexico','GDP in Trillions of U.S. Dollars','Life Expectancy At Birth (year)')

plt.subplot(3,2,5)
plotBestFit(US,'GDP','LE','m','GDP vs LE for USA','GDP in Trillions of U.S. Dollars','Life Expectancy At Birth (year)')

plt.subplot(3,2,6)
plotBestFit(ZW,'GDP','LE','k','GDP vs LE for Zimbabwe','GDP in Trillions of U.S. Dollars','Life Expectancy At Birth (year)')
plt.show()
```

**Calculate correlation coefficents.**

```python
corCoef = {'Chile':round(np.corrcoef(CL['GDP'], CL['LE'])[0, 1],3),'China':round(np.corrcoef(CN['GDP'], CN['LE'])[0, 1],3),'Germany':round(np.corrcoef(DE['GDP'], DE['LE'])[0, 1],3),'Mexico':round(np.corrcoef(MX['GDP'], MX['LE'])[0, 1],3),'USA':round(np.corrcoef(US['GDP'], US['LE'])[0, 1],3),'Zimbabwe':round(np.corrcoef(ZW['GDP'], ZW['LE'])[0, 1],3)}

df = pd.DataFrame(corCoef.items(), columns=['Country', 'corCoef'])
print(df)
```

# Conclusions

**Life Expectancy**

As seen in the figure below, all countries showed a trend towards increased in life expectancy over time. 

Things to note:
- Zimbabwe life expectancy didn't begin to increase until 2004.
- Chile and Mexico both showed decreses in life expectancy after 2007, but the positive trend returned after 2010.



**Gross Domestic Product**

As seen in the figure below, GDP generally increased over time for all countries.

Things to note:
- Zimbabwe didn't see an increase in GDP until 2008 when the GDP switched from a negative trend to increasing rapidly.
- Chile's GDP began to decrease after 2013.
- Germany's GDP rate of increase was positve until 2008. After 2008, the GDP fluctuated each year with resulting in only a small increase between 2008 and 2014.
- Chile, Mexico, and USA showed a decrease in GDP between 2008-2009, but GDP incresed between 2009-2010.




**Corrolation between life expectancy and GDP.**

The figure below shows scatter plots of GDP vs life expectancy. The black line in each plots represents a linear best fit line. For each country, the slope of the best fit line is positive which suggests a posive correlation between GDP and life expectancy.

This relationship is further seen by calculating the correlection coefficent of GDP and life expectancy. The correlection coefficent for each country is listed below.

- Chile - 0.950 
- China - 0.909 
- Germany - 0.933 
- Mexico - 0.932 
- USA - 0.982 
- Zimbabwe - 0.966 

All the correlection coefficents are postive and near 1, which indicates that there is a strong positive correlection between GDP and life expectancy.
